# Week 0 — ☺️ Introduction to Python


To do computer programming, first need a way to communicate with the machine.


Computers don't understand human languages like English or Mandarin - they only understand binary logic: 1s and 0s, or more precisely, high and low voltages flowing through hardware.
So, early computer scientists came up with a clever solution: assign a unique binary code to every letter, number, and punctuation mark. For example:

- A → 01000001 (65 in decimal)
- B → 01000010 (66 in decimal)
- C → 01000011 (67 in decimal)

 and so on...

Every time you tap the letter 'A' on your keyboard, it gets encoded into that binary sequence. We call this the ASCII standard, and it's what allows your keystrokes to travel through the hardware and show up on your screen. These strings of 1s and 0s is the language that computers understand.

But if we want the computer to actually do something - run a calculation, follow a set of instructions - writing everything out in raw binary would be too much work!. So computer engineers designed more human-friendly programming languages that handle the translation for you. When you type a command, the language takes care of converting it into the low-level signals the machine understands.

<img src="images/lec00_language_hierarchy.jpg" width="500">

This started in the early days of computing (~1960's), but now scientists have developed more and more user-friendly, human-like (high-level) languages that can have multiple translation layers, which makes computer programming easier for us humans.

We've also expanded beyond ASCII. The UTF-8 standard now covers non-Latin scripts too - Mandarin, Japanese, Greek, Arabic, and many more - so keyboards and screens around the world can all be translated into the same underlying binary language.

## Python          <img src="images/Python_logo.jpg" width="40">

Python was first released in the early 1990s but really started gaining traction around 2010. Since 2021 it has held the top spot as the world's most popular programming language ([TIOBE index](https://www.tiobe.com/tiobe-index/)).

So what makes Python so popular?
- It reads like plain English. Python code is clean, minimal, and beginner-friendly. You spend more time thinking about the problem and less time wrestling with the language.
- It's completely free. Python is fully open-source - anyone can download and use it at no cost.
- There's a library for almost everything. Chances are, whatever you want to do, someone has already built a tool for it. These ready-made tools are packaged into libraries (also called packages) that you can plug straight into your code. There are currently over 600,000 of them available on [PyPI](https://pypi.org/), Python's main package repository.
- It's incredibly versatile. Web development, scientific modelling, machine learning, office automation, database management, cybersecurity - Python handles all of it, and integrates smoothly with many other languages and software tools.
- It has serious backing. Major tech companies like Google and Meta rely heavily on Python, and millions of developers worldwide use and contribute to it daily. This means the language and its ecosystem keep growing and improving constantly.

⏬ The one downside for everyday users like you and me, is package conflicts. With so many libraries available, and new versions being released all the time, it's not unusual for two packages in the same project to be incompatible with each other. Frustrating, but very common.😒

😁 The good news: there's a clean solution called a virtual environment, which keeps each project's packages neatly separated. We'll set one up shortly.

## Creating a virtual environment

<img src="images/lec00_Anaconda_Base.jpg" width="400">

The "(base)" indicates you are in your computers base - default location. Although there's nothing wrong with installing packages and running your scripts in the base, it is not good pracice. In the long-term you might want to build different python projects with a different set of packages or versions of packages - and if you have all of them stored in the (base) you might run into conflicts! 
So, to avoid that it is a good practice to create a virtual environment and keep everything related to a specific project within that environment. Think of this like a 'folder' on your computer. 
So, we will create a virtual environment for this course MLE5217. And in the next semester if you study another python course, you can create another virtual environment in its name.

to create a virtual environment, type the following command (**One time action**) in Anaconda prompt(or Terminal) - and press Enter 


```python
conda create -n mle5217 python=3.10
```

<img src="images/lec00_Anaconda create venv.jpg" width="400">

Note that: mle5217 is the name (-n) of the environment and 3.10 is the version of python that we are installing in this environment.

This is a one-time action - once the environment is created, you don't have to run this command again - it is there on your computer. 

Wait for the computer to finish this task. When it's done it'll be ready to take your input again.

To access it you have to **activate it everytime**. Type in the following and Enter.


```python
conda activate mle5217
```

<img src="images/lec00_Anaconda activate venv.jpg" width="400">

Note: how (base) is now replaced by (mle5217) - that means you are inside the virtual environment

Now we need to install a couple of things - since we are using jupyter notebook as our IDE (integrated development environment) which is like the user-interface: we need to install Jupyter.


```python
conda install jupyter
```

<img src="images/lec00_Anaconda install jupyter.jpg" width="400">

Then we need to install something called a 'Kernel' - a Kernel in Python is what helps us Run our code.


```python
conda install ipykernel
```

<img src="images/lec00_Anaconda install kernel.jpg" width="400">

Register the Kernel name


```python
python -m ipykernel install --user --name=mle5217 --display-name "Python (mle5217)"
```

# How do Machines learn?

To understand how machines learn, let's start with something we all know - how we learned things as babies.

## How you learned 

Imagine your parent is taking you for a walk in your stroller 👶, and a dog walks by 🐶. Your parent points and says "Dog!" You look at it - fur, four legs, a tail, floppy ears - and your brain labels it: that's a dog.
A little further down the road, a cat crosses your path 🐱. It also has fur, four legs, a tail, and ears. So naturally, you think: dog! But your parent shakes their head - "Cat."
Hmm. You think about it. Maybe it's the size? The cat is smaller. So you update your mental model: same features, small → cat, big → dog.
Then one day you meet a tiny chihuahua. Your parent says "Dog." Your brain goes back to the drawing board. Size isn't the answer after all.

So you keep observing. More cats. More dogs. Each time, you pick up on subtler and subtler differences - the shape of the face, the way the ears sit, how they move. No single feature tells the whole story, but taken together, all these small details let you tell them apart reliably.
Eventually, you can look at a photo of a breed you've never seen in your life and still say - confidently - cat or dog.

But here's the funny part: if someone asks you to explain exactly how you tell them apart, it's surprisingly hard to put into words. The model in your head is complex, built up from hundreds of examples over time. 🧠

## 💻 How Machines Learn

Machine learning works in a remarkably similar way.
We feed the machine lots of data - thousands of images of cats and dogs - and ask it to figure out the difference. With only a handful of images, the model it builds is weak and unreliable. But the more examples it sees, the more it refines a complex mathematical model of what a cat looks like and what a dog looks like - not based on one simple rule like size or colour, but on a rich combination of subtle patterns and features. With that model, it can then correctly classify a brand new image it has never seen before.

There is one important difference though. Humans are born curious - we naturally observe, analyse, and build mental models without being told to. Computers don't have that built in. The hardware alone can't look at an image and figure out what's in it.
That's where we come in.

Machine learning is about us - humans - giving the machine the right tools and techniques so that it can learn to do a task on its own. We provide the data, we choose the model, and we set up the process. The machine does the learning.




## Machine Learning Vs Computer Programming

Machine Learning vs Traditional Programming
Computers cannot think. They need to be told exactly what to do -

🐾 step by step, no exceptions.

When you see a dog, your brain recognises it in a fraction of a second - effortlessly and automatically. Now try to write a traditional computer programme to do the same thing. Suddenly, what feels trivial becomes nearly impossible. How would you even begin? A tangle of if/else statements checking for fur? Four legs? A tail? And what happens when a cat ticks all the same boxes?

*The Traditional Way — Rules First, Data Second*

Traditional programming works like classroom learning. Your teacher explains the rules of multiplication first, then hands you a worksheet full of problems to practise on. The rules come first, the data comes second.

In the same way, a traditional programme is built on rules that a human has written out in advance. The computer just applies them.

*The Machine Learning Way — Data First, Rules Emerge*

Machine learning flips this completely. Instead of defining the rules upfront - which we often can't, because we don't know what they are - we give the machine a large amount of data and ask it to find the patterns hidden inside. Those patterns, hopefully, reveal the underlying rules on their own.

This is actually quite similar to how scientific discoveries are made. Take astronomy as an example. For centuries, scientists carefully recorded the positions of stars and planets night after night. When they analysed all those numbers together, patterns emerged - and from those patterns, they were able to work out how the Earth and other objects move through space. The rules of planetary motion weren't written first; they were discovered from the data.

So What Does the Machine Actually Need?
Data alone isn't enough. To find patterns in data, you also need to tell the machine how to look for them. That's where mathematics comes in - specifically, statistical and mathematical methods that guide the search. We call these machine learning models.
This is why machine learning sits at the intersection of three fields:

**Statistics** — provides the concepts and techniques for analysing data and finding patterns

**Linear algebra** — provides the mathematical framework for carrying out those calculations efficiently

**Programming** — translates all of the above into something a computer can actually execute

## Types of Machine Learning

We said that to help a machine learn, we need to provide it with mathematical and statistical methods to guide the pattern search. But before picking a method, we need to ask a more fundamental question:
What kind of problem are we actually trying to solve?
Because it turns out, not all learning looks the same. Just like how you'd study differently for a maths exam versus learning to ride a bike, different problems call for different types of learning. Here's the landscape:


| Type | Short description |
|---|---|
| Supervised learning | Learn from labelled examples |
| Unsupervised learning | Find hidden patterns, no labels |
| Semi-supervised learning | A little labelled, a lot unlabelled |
| Self-supervised learning | Create your own labels from the data |
| Reinforcement learning | Learn by trial, error, and reward |


Let's go through each one. 👇

## 🏷️ Supervised Learning

Supervised learning is the most straightforward type - and the one we'll spend most of our time on in this course.

Remember the baby-and-dog example from earlier? Your parent pointed at a dog and said "Dog!", then pointed at a cat and said "Cat!" That pointing - that labelling - is exactly what makes it supervised. Someone has already done the hard work of labelling the data for you, and the machine learns by studying those labelled examples.

<img src="images/lec01_categories.png" width="400">

In more formal terms: you give the machine a dataset where every data point has an input (e.g. an image, or a set of material properties) and a known output (e.g. "cat", or a measured bandgap value). The machine finds the relationship between input and output, and uses it to predict the output for new, unseen inputs.
Supervised learning splits into two flavours depending on what you're predicting:

**Classification** - predicting a category. Is this material a metal, semiconductor, or insulator? Is this image a cat or a dog?

**Regression** - predicting a number. What is the bandgap of this new compound? What will the melting point be?

Both will feature heavily in this course. 🎯

## 🔍 Unsupervised Learning

If supervised learning is learning with a teacher, unsupervised learning is exploring without one - no labels, no right answers, just raw data.

Here's a good way to picture it. Imagine you arrive at a new job at a library, and someone drops a massive pile of unsorted books at your feet and says: "Sort these out." No further instructions. How do you do it?
You might group by subject (chemistry, history, fiction...), or by reading level (children vs adults), or by size (big books on the bottom shelf), or alphabetically by author. There's no single correct answer - the best grouping depends on what the library actually needs.

<img src="images/lec01_books.png" width="300">

Unsupervised learning works the same way. You hand the machine a pile of unlabelled data and ask it to find structure - groupings, patterns, relationships - on its own. The two most common tasks are:

**Clustering** - grouping similar data points together (like sorting the books by subject). A common example in materials science: grouping compounds with similar properties even when you don't know their crystal structure in advance.

**Dimensionality reduction** - simplifying data that has many, many features down to a smaller set that still captures the important variation. Useful when you have hundreds of material descriptors and need to visualise or compress them.

## 🤝 Semi-Supervised Learning

Here's a very practical problem: labelling data is expensive. It takes time, money, and expertise. In materials science, measuring a property experimentally for thousands of compounds is simply not feasible.
Semi-supervised learning is the pragmatic middle ground. You have a small amount of labelled data and a large amount of unlabelled data, and you use both together. The idea is that even without labels, the structure of the unlabelled data carries useful information - and a clever algorithm can exploit that structure to make better use of the few labels you do have.

Think of it like this: you are putting together a mechanical instrument. You have a instruction manual but the information is limited. So you use that information but at the same time you look at components and work out how and where they fit. So, you are using 'labeled' data as well as 'unlabaled' data.

<img src="images/lec01_semisupervised.png" width="300">

Semi-supervised learning is increasingly common in real-world ML applications where labelled data is scarce. 🏗️


## 🤖 Self-Supervised Learning

Self-supervised learning is a clever twist: instead of relying on human-provided labels, the machine generates its own labels from the data itself.

A famous example: large language models like the ones that power modern AI chatbots are trained by hiding a word in a sentence and asking the model to predict what the missing word is. The "label" (the correct word) comes directly from the text itself - no human labelling required. This lets these models train on enormous amounts of data essentially for free.

<img src="images/lec01_selfsupervised.png" width="500">

In materials science, a self-supervised approach might involve hiding part of a crystal structure and asking the model to predict the missing atoms - allowing it to learn rich representations of materials without needing expensive property labels.

Self-supervised learning has become one of the most exciting and fast-moving areas of modern ML research. 🚀

## 🎮 Reinforcement Learning

Reinforcement learning is the most different of the bunch - and honestly, the most fun to think about.
Forget data tables and labels. In reinforcement learning, a agent (the learner) interacts with an environment and learns by trying things out. When it does something good, it gets a reward. When it does something bad, it gets a penalty. Over many, many attempts, it figures out which actions lead to the best outcomes.

Think of it like playing a video game 👽 - not by showing the path to win the level, but by giving you a point when you win, and reducing a life when you lose 💔. You figure out the rules & tricks through experience.

<img src="images/lec01_supermario.png" width="300">

Some famous examples:

- AlphaGo (DeepMind) - learned to play the board game Go at superhuman level purely by playing millions of games against itself.
- Robot arms - learn to pick up and manipulate objects through trial and error in simulation.
- Materials discovery - reinforcement learning has been used to guide the search for new materials by iteratively suggesting candidates and updating based on experimental feedback.

The key ingredients are:

- An agent - the learner/decision maker
- An environment - what the agent interacts with
- Actions - what the agent can do
- Rewards - feedback signals that tell the agent how well it's doing
- A policy - the strategy the agent develops over time

Reinforcement learning is powerful but also complex to set up - it's less commonly used in standard materials informatics than supervised or unsupervised learning, but it's worth knowing it exists. 🎯

## Why do we need to understand whether a given task is Supervised/ Unsupervised / Reinforced etc.?

Because it directly influences how we design our solution - including the type of data we need, the model we choose, and the learning strategy we apply.

1. It defines the structure of the learning problem

    Supervised Learning → Labeled data, so the task becomes learning a mapping from inputs to outputs.

    Unsupervised Learning → No labels, so the task is about finding structure or patterns in the data.

    Reinforcement Learning → Learning happens through interaction with an environment over time, where  rewards guide behavior.

👉 Knowing the task type tells you what form of feedback your model will use to learn.


2. It determines the choice of algorithms and tools
   Supervised? → Use classification or regression models (e.g., decision trees, SVMs, neural networks).

   Unsupervised? → Use clustering, dimensionality reduction, or anomaly detection methods.

   Reinforcement? → Use Q-learning, policy gradients, or actor-critic methods.

   👉 Each paradigm has its own set of algorithms and libraries.

3. It affects data requirements and processing
   Supervised → We need a large, labeled dataset, and label quality heavily influences performance.

   Unsupervised → We need diverse and rich feature sets; interpretability can be harder.

   Reinforcement → We often need a simulator or interactive environment and must handle delayed rewards.

   👉 Mismatch between task and approach leads to poor performance and wasted resources.

4. It helps set realistic expectations

   Supervised learning is typically easier to train and evaluate but requires labeled data.

   Unsupervised learning can uncover hidden patterns, but it's harder to validate.

   Reinforcement learning can achieve superhuman performance, but often needs massive computation and trial-and-error.

    👉 Understanding the task type helps communicate what’s feasible and how long development might take.


## The Machine Learning Workflow

So far we've talked about what machine learning is. Now let's talk about how you actually do it.
In practice, machine learning is not just about picking a model and hitting "run". It's a structured process - and skipping steps is one of the fastest ways to end up with a model that looks great on paper but falls apart in the real world.

There is no single universal solution that works for every problem. But there is a general workflow that practitioners follow to make sure the model they build is reliable, mathematically valid, and actually useful.

Here's the big picture:


<img src="images/lec01_MLprocess.png" width="400">


**1. 🎯 Problem Definition**

Everything starts here. Before touching any data or writing any code, you need to clearly understand what you are trying to solve:

What is the question? (e.g. "Can we predict the bandgap of a new compound?")

What type of learning does it require? (supervised, unsupervised, reinforcement...)

If supervised - is it a regression task (predicting a number) or a classification task (predicting a category)?

What does success look like? (How accurate is accurate enough?)

What resources are available? (How much data? How much compute?)

Getting this step right saves an enormous amount of time later. A poorly defined problem leads to a perfectly optimised answer to the wrong question. 😬



**2. 📦 Data Collection & Understanding**

Once you know what you're solving, you need data. This step involves:

Collecting or sourcing a suitable dataset

Understanding what each column represents

Checking the size of the dataset - do you have enough data for the complexity of the problem?

Getting a first feel for the data: distributions, ranges, obvious outliers

In materials science, data might come from experimental measurements, computational simulations (e.g. DFT calculations), or public databases such as the Materials Project or AFLOW.


**3. 🧹 Data Preprocessing**

Real-world data is almost never clean and ready to use straight away. This step is often the most time-consuming - and the most important. It includes:

Handling missing values - drop them, or fill them in intelligently

Removing or investigating outliers - remembering that in materials science, an outlier might be the most interesting data point!

Feature engineering - transforming raw data into meaningful inputs for the model (e.g. converting a chemical formula into numerical descriptors)

Scaling & normalisation - many ML models are sensitive to the range of input values, so features often need to be rescaled

Splitting the data - dividing into a training set (what the model learns from) and a test set (what we use to evaluate it on unseen data)

*The golden rule: the model must never see the test set during training. This is how we simulate real-world performance.*


**4. 🤖 Model Selection & Training**

Now the model comes in. Based on your problem definition, you choose an appropriate algorithm - a linear regression, a decision tree, a support vector machine, and so on. You'll meet all of these in this course.
Training is the process of feeding the training data into the model and letting it find the patterns. Mathematically, this usually means finding the set of internal parameters that minimise the difference between the model's predictions and the known correct answers.


**5. 📊 Model Evaluation**

Once trained, you evaluate the model on the test set - data it has never seen before. This gives you an honest estimate of how well it will perform in the real world.
Different tasks use different metrics:

Regression: Mean Absolute Error (MAE), Root Mean Squared Error (RMSE), R² score
Classification: Accuracy, Precision, Recall, F1 score, ROC curve

A model that performs well on training data but poorly on test data is overfitting - it has memorised the training examples rather than learning the underlying pattern. Spotting and fixing overfitting is one of the central challenges of machine learning, and we'll cover it in depth later in the course.

**6. 🔧 Model Tuning & Improvement**

Rarely does the first model you try give the best result. This step involves:

Adjusting hyperparameters - settings that control how the model learns (distinct from the internal parameters the model learns itself)

Trying different algorithms and comparing results

Going back to preprocessing if the data quality is limiting performance

Using techniques like cross-validation to get a more robust estimate of model performance

This step is often iterative - you go around the loop several times before settling on a final model.

**7. 🚀 Deployment & Interpretation**

Finally, once you're happy with the model, it gets put to use - making predictions on new, real-world data. In a research context, this might mean using the model to screen thousands of candidate materials for a desired property, shortlisting the most promising ones for experimental synthesis.

In many scientific applications, it's also important to interpret the model - to understand which features it found most important, and whether its reasoning makes physical sense. A model that gives the right answer for the wrong reasons is dangerous in science. 🔬

**🔄The Workflow is Rarely Linear🔄**

One important thing to know: in practice, this process loops back on itself constantly. You evaluate the model and realise the data needs more cleaning. You try a new algorithm and go back to redefine the problem slightly. You interpret the results and discover a data collection issue.

This is normal - and expected. The workflow is a guide, not a rigid recipe. 🔄

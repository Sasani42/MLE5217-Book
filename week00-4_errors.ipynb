#  ❌ A Beginner's Guide to Errors

First, the good news: errors are normal. Every programmer, beginner or expert, sees them every day. Python is not broken, and neither are you. Python is just being very literal - it does exactly what you tell it to do, and when something doesn't make sense, it stops and tells you why.

Think of Python as a very helpful but very fussy assistant. 🤖

## 📌 How to Read an Error Message
When something goes wrong, Python gives you a traceback — a message that tells you:

Where the problem is (the line number)
What the problem is (the error type)

Always read the last line first — that's where the actual error is described.

🔴 **The Most Common Errors**

-----------------------------------------------------------------------------------------------

<span style="color: red;"> **SyntaxError**</span> — "I can't even read this"

Python cannot understand the code. Usually a missing bracket, colon, or quote.

```python
print("hello"   # forgot to close the bracket
```

💡 Fix: check for missing ), :, or "

-----------------------------------------------------------------------------------------------

<span style="color: red;"> **NameError**</span> — "I don't know what that is"

You used a variable that Python hasn't seen yet — maybe a typo, or you forgot to run a previous cell.

```python
print(result)   # result was never defined
```

💡 Fix: check the spelling, and make sure you ran the cell that creates the variable.

-----------------------------------------------------------------------------------------------

<span style="color: red;"> **TypeError</span>** — "That doesn't make sense"

You tried to do something with the wrong type of data — like adding a word and a number.

```python
"bandgap" + 1.12   # a word plus a number?
```

💡 Fix: check what types you're working with. Use type(x) if unsure.

-----------------------------------------------------------------------------------------------

<span style="color: red;"> **ValueError** </span>— "That value makes no sense here"

The type is fine, but the value itself is impossible for that operation.

```python
int("hello")   # hello is a string, but not a number-like one
```

💡 Fix: check what values your function actually expects.

-----------------------------------------------------------------------------------------------

<span style="color: red;"> **IndexError**</span> — "That position doesn't exist"

You asked for item number 5 in a list that only has 3 items. Remember — Python starts counting from 0!

```python
elements = ["Fe", "Cu", "Au"]
print(elements[5])   # there is no position 5!
```

💡 Fix: remember indexing starts at 0. The last item is always at position len(list) - 1.

-----------------------------------------------------------------------------------------------

<span style="color: red;"> **KeyError** </span> — "I don't have that column / key"

You asked for a column or dictionary key that doesn't exist — usually a typo.

```python
import pandas as pd
data = {'material': ['Silicon', 'Gallium Arsenide', 'Diamond'],
    'bandgap': [1.12, 1.42, 5.47]
}

df = pd.DataFrame(data)

df["bandgipp"]   # did you mean "bandgap"?
```

💡 Fix: check your spelling. Use df.columns to see the exact column names.

-----------------------------------------------------------------------------------------------

<span style="color: red;"> **ZeroDivisionError** </span>— "You can't divide by zero!"

Mathematically impossible — Python agrees with your maths teacher on this one. 😄

```python
mass=10
vol=0
density = mass / vol
```

💡 Fix: check where your zero is coming from — often a missing or empty value in the data.

-----------------------------------------------------------------------------------------------

<span style="color: red;"> **ImportError / ModuleNotFoundError** </span>— "I've never heard of that package"

The package is either not installed, or you have a typo in the name.

```python
import numppy   # one p too many!
```

-----------------------------------------------------------------------------------------------

<span style="color: red;"> **AttributeError**</span> — "That object can't do that"

You called a method that doesn't exist for that type of object — like asking a number to .append() something.

```python
"hello".append("world")   # strings don't have append!
```

💡 Fix: check what methods are available for your object type.

-----------------------------------------------------------------------------------------------

<span style="color: red;"> **FileNotFoundError** </span>— "I looked everywhere, that file isn't there"

Python can't find the file at the path you gave it — usually a typo in the filename or wrong folder.

```python
pd.read_csv("data/materails.csv")   # typo in filename
```

💡 Fix: double-check the filename and folder. Copy-paste the path rather than typing it.

🛠️ General Tips

- Read the last line of the error first — that's the actual problem.
- Look at the line number — Python tells you exactly where to look.
- If you can't fix it yourself, Copy the error message and search online — you are almost certainly not the first person to see it, or ask a Chatbot (note that sometimes Chatbots give wrong answers).
- Don't panic — an error is just Python asking for a small correction. Fix it and move on! 😊

## Loops ➿

A lot of the time there are tasks that we need to do multiple times in a job.

For example if we need to calculate the GPA of our class of 200 students, we can write a function to do that.
It would be something like 

$$\frac{(G_1*U_1 + G_2*U_2 +G_3*U_3+....+G_N*U_N)}{(U_1+U_2+...+U_N)}$$

Where $G_n$ is the grade and $U_n$ is the number of units. 
Then we can convert this into a python code, but do we repeat the same code 200 times to calculate for each student?

No! this is a waste of time and computation. We have better ways of dealing with this type of iterative tasks. They are called loops.

## Functions ( )

A function carries out a specific task. Remember print("Hello!")?

Well, print() here is a function. You can always recognize a function by the brackets (). Sometimes there can be something written inside the brackets and sometimes it maybe empty, either way it is a function and it has been defined to do a specific task.

function
- a reusable block of organized code designed to perform a single, specific task.


print() is an <span style="color: blue;"> in-built python function</span>, meaning that it comes with your original python package. And python has coded the specific sequence of tasks that should be carried out (from taking the string that you have input "Hello!" and making it appear in the output) whenever a user executes a print() function. 

The code that makes print() work is not visible to us, it is hidden (encapsulated). But whenever you run print("Hello!"), python interpreter finds this code and executes the task.

Apart from print() there are a few other in-built python functions, some of which you have already used. You can find the full list [here](https://docs.python.org/3/library/functions.html).

Additionally,there are functions that come with python packages: <span style="color:blue;">standard library or third-party functions</span>. A simple package that you have already used is math. And you will learn about other important packages soon.

Whenever you need to use a function from a package, you cannot simply call it like you called print(). You also need to specify which package the function belongs to. 

e.g. to use sqrt() function you should write as "math.sqrt()"

to use array() function, you should write as "numpy.array()"

Thirdly, you can also <span style="color:blue;">define your own function</span>. Here, you have to write out the code for the function to tell python what to do whenever it encounters your user-defined function.

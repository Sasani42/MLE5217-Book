## Let's see how this works Mathamatically

<img src="images/lec03_lss.png" width="500">



The residuals (gaps between each datapoint $y_i$ and the corresponding point on the line $y_{li}$) are:

$(y_1 - y_{l1}) , (y_2 - y_{l2}) , (y_3 - y_{l3}) \dots , (y_n - y_{ln})$

To find the sum of squares - we will square each of these terms and add them up

residual sum of squares $= (y_1 - y_{l1})^2 + (y_2 - y_{l2})^2 +(y_3 - y_{l3})^2 +\dots + (y_n - y_{ln})^2$ 

Which we can write in a compact form

residual sum of squares $= \sum_{i=1}^{n}(y_i - y_{li})^2$

And then normalize it based on the number of datapoints

$\Delta = \frac{1}{n}\sum_{i=1}^{n}(y_i - y_{li})^2$

The equation of any straight line can be written as

$y=a+bx$

where $b$ is the gradient (slope) and $a$ is the intercept

So, all the possible lines through the dataset can be represented with this equation, by varying $a$ and $b$.

if we take datapoints on one of these line $(y_{li})$:

$y_{li}=a_i + b_i x$

So, our above equation can be written as 

$\Delta = \frac{1}{n}\sum_{i=1}^{n}(y_i - (a_i+b_i x))^2$

$n$ is the number of datapoints
$x_i, y_i$ are the data points.

**To find the line of best-fit**, we have to find the values of $a_i$ and $b_i$ such that it minimizes the value of $\Delta$

```{admonition} Extra Math 
:class: dropdown

Mathamatically the analytical solution to the least-squares problem is found by partial differentiation of the equation by $a$ and $b$ and finding the solutions where the expression is zero. 

$\frac{\partial\Delta}{\partial b}=0$

and

$\frac{\partial\Delta}{\partial a}=0$

After multiple steps of differentiation (changing summation limits, and re-arranging) leads us to the following solutions for $a$ and $b$

$b=\frac{n\sum x_i y_i-\sum x_i \sum y_i}{n\sum x_i^2-(\sum x_i)^2}$

$a=\frac{\sum x_i^2 \sum y_i - \sum x_i \sum x_i y_i}{n\sum x_i^2-(\sum x_i)^2}$


Which means if you know the datapoints, you can calculate the exact equation for the line of best-fit. 

Note: This is the analytical solution for Linear regression. However, how we calculate mathematical solutions is different to how computers handle them. To implement this in Python we need to convert the math into a linear algebra framework. 

[See the matrix supplement for further detail](./week03-Regression_Supplementary.ipynb)


```

```{code-cell} python3

#Import the necessary packages
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

data=pd.read_csv("3_simple_regression.csv") #Read the data and save it as 'data'

```

````python
#Let's plot the data to see if that looks correct
data.plot(kind="scatter",x="Carbon_Content",y="Hardness")
````

````python
#To fit a Regression line, we need to import the package from sci-kit learn
from sklearn.linear_model import LinearRegression
model=LinearRegression() #making an instance of the class
````

````python
X=data[["Carbon_Content"]] #I'm assiging the E-field column values to the variable X
y=data[["Hardness"]] #and Strain column to the variable y
````

````python
result=model.fit(X,y) #Fittin the model to X,y data and saving the fit as result
````

````python
y_pred=model.predict(X) #Using the model to predict what y should be, based on X
````

````python
#Visualizing prediction
plt.scatter(X,y)
plt.plot(X,y_pred,color='red') 
plt.show()
````

````python
model.coef_  #gradient of the model
````

````python
model.intercept_  #Intercept of the model
````

````python
r2=model.score(X,y)       #r2 value of the model
print(r2)    
````

If a new steel sample has **0.35 wt% carbon**, what hardness would you *guess* from your line?

````python
#Extra
y_35=model.predict(np.array([[0.35]]))  # array needs to take two dimensions [[]], and predict takes an array as argument
print(y_35)
````

Do you think that the hardness prediction for 0.35 wt% carbon is reliable?

**Why??**

```{admonition} Interpolation & Extrapolation
:class: tip dropdown
Yes - fairly reliable! Predicting *within* the range of carbon values the model was trained on is called **interpolation**, and it's usually on safe ground, since the model has already seen how hardness behaves nearby.

Predicting *outside* that range is **extrapolation** - much riskier, because the model is guessing where it has no data, and the real trend might curve or flatten out there.

Since 0.35 wt% carbon sits comfortably inside our data, this is interpolation, so we can lean on the prediction with reasonable confidence. 👍
```

# A Gentle Introduction to Linear Algebra

If the words linear algebra make you nervous - don't worry. You've probably already been doing bits of it without realising it. And by the end of this section, you'll see exactly why it shows up in machine learning, and what parts of it actually matter for this course.

----------------------------------------------------------------------------

**🤔 What Even Is Linear Algebra?**

At its heart, linear algebra is the mathematics of organised numbers - how to store them, manipulate them, and extract meaning from them efficiently.
Think about it this way. You have a dataset of 1,000 materials, each described by 50 different properties (bandgap, density, melting point, atomic radius...). That's 50,000 numbers. How do you store all of that neatly? How do you run calculations on all of it at once without writing 50,000 lines of code?
Linear algebra gives us the tools. 🧰

----------------------------------------------------------------------------

## 📦 Scalars, Vectors, and Matrices

Everything in linear algebra builds from three simple ideas:

**Scalar** - just a number. A single value. Nothing new here.

$x = 1.12 \text{ (the bandgap of Silicon, in eV)}$

**Vector** - a list of numbers

An ordered collection of numbers, written as a column (or a row). Think of it as describing one thing using multiple numbers.

$\mathbf{x} = \begin{bmatrix} 1.12 \\ 2.33 \\ 1414 \end{bmatrix}$

This could represent one material - Silicon - described by its bandgap, density, and melting point. In machine learning, we call this a feature vector. Every material in your dataset gets its own vector.

Vectors have a length (also called *magnitude*), which tells you how "big" the vector is overall, and a *direction*, which captures the pattern of values inside it. Both of these turn out to be useful in ML.

If you imagine the 3 numbers being represented in 3-dimensional space along $x,y,z$ axes:

$x = 1.12,  y=2.33, z=1414$ 

Then Silicon will appear as a point in that space. The direction of that point from the origin $(0,0,0)$ will be the direction of the vector. And you can calculate the length of that vector using the distance between the point of Siliconand the origin.


**Matrix - a table of numbers**

Stack several vectors together and you get a matrix - a rectangular grid of numbers with rows and columns.

$\mathbf{X} = \begin{bmatrix} 1.12 & 2.33 & 1414 \\ 0.67 & 5.32 & 938 \\ 1.42 & 8.96 & 1357 \end{bmatrix}$

Here, each row is one material, and each column is one property. This is your entire dataset, packaged into a single mathematical object. In ML, this is called the feature matrix, and almost every algorithm starts here.

A matrix with $m$ rows and $n$ columns is said to have shape $(m \times n)$ - you'll see this notation constantly when working with NumPy and pandas.

**Tensor** - the next level up

A tensor is simply a generalisation - a scalar is a 0-dimensional tensor, a vector is 1-dimensional, a matrix is 2-dimensional, and a tensor can be 3-dimensional or higher. You'll encounter tensors more in deep learning (a future course), but it's good to know the word. 🗂️

## ➕ Operations on Vectors and Matrices

Now that we can store data neatly, we need to do things with it.

**Vector Addition & Scalar Multiplication**

These work exactly as you'd expect - element by element:

$\begin{bmatrix} 1 \ 2 \ 3 \end{bmatrix} + \begin{bmatrix} 4 \ 5 \ 6 \end{bmatrix} = \begin{bmatrix} 5 \ 7 \ 9 \end{bmatrix}$

Subtraction works the same as addition.

$\qquad 3 \times \begin{bmatrix} 1 \ 2 \ 3 \end{bmatrix} = \begin{bmatrix} 3 \ 6 \ 9 \end{bmatrix}$


**The Dot Product** - where things get interesting

The dot product takes two vectors and returns a single number:

$\mathbf{a} \cdot \mathbf{b} = a_1 b_1 + a_2 b_2 + a_3 b_3 + \ldots$

For example:

$\begin{bmatrix} 1 \ 2 \ 3 \end{bmatrix} \cdot \begin{bmatrix} 4 \ 5 \ 6 \end{bmatrix} = (1 \times 4) + (2 \times 5) + (3 \times 6) = 4 + 10 + 18 = 32$

Why does this matter? 

Later on in this course, you might see that the dot product measures how similar two vectors are in direction. If two vectors point in the same direction, the dot product is large. If they are perpendicular (completely unrelated), the dot product is zero. This idea of measuring similarity is fundamental to many ML algorithms.


**Matrix Multiplication**

This is the big one. Multiplying two matrices together combines their information in a structured way.

$\mathbf{C} = \mathbf{A} \times \mathbf{B}$

Each element of the result is a dot product between a row of $\mathbf{A}$
and a column of $\mathbf{B}$. It sounds fiddly, but the key thing to know is:

For $\mathbf{A} \times \mathbf{B}$ to work, the number of columns in $\mathbf{A}$ must equal the number of rows in $\mathbf{B}$.

The result has the same number of rows as $\mathbf{A}$ and the same number of columns as $\mathbf{B}$.

In ML, matrix multiplication is how we apply a model to an entire dataset in one shot - instead of looping through each sample one by one. NumPy does this for you with a single line of code, but now you know kind of what's happening underneath.
